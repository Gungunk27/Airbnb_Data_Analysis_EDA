```python
df=data[data['number_of_reviews']<750]
sns.boxplot(data=df,x='number_of_reviews')
```

```python
# Number of Reviews 
plt.figure(figsize=(6,6))
sns.histplot(data=df, x='number_of_reviews', bins=50)
plt.title("Number of Reviews")
plt.ylabel("Frequency")
plt.show()
```

```python
# Availability_365
plt.figure(figsize=(6,6))
sns.histplot(data=df, x='availability_365', bins=50)
plt.title("Availability_365")
plt.ylabel("Frequency")
plt.show()
```

```python
availability_by_neighborhood = df.groupby('neighbourhood_group')['availability_365'].mean()
print(availability_by_neighborhood)
availability_by_neighborhood.plot(kind='bar', figsize=(8, 6), color='skyblue')
plt.title('Average Availability by Neighborhood Group')
plt.xlabel('Neighborhood Group')
plt.ylabel('Average Availability (Days)')
plt.show()
```
